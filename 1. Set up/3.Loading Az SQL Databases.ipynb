```python
hos_a_jdbc_url = os.getenv("HOSA_JDBC_URL")
hos_b_jdbc_url = os.getenv("HOSB_JDBC_URL")
```

```python
connection_props = os.getenv("CONNECTION_PROPS")
```

```python
hos_a_dept_df.write \
    .mode("overwrite") \
    .jdbc(url=hos_a_jdbc_url, table="dbo.departments", properties=connection_props)
```
